```python
#Created a function to check if '\n' is present in any field
def check_multiple_rows(a):
    if '\n' in a:
        return True
    return False
#Created a feature which would return True if '\n' is present in the feature name
df['multiple_rows_present']=df['name'].apply(check_multiple_rows)
```

```python
#Unnesting of two rows mentioned above
df_1_list = []
for i in range(1, len(df.iloc[112845,1].split('\n'))):
    df_1_list.append(df.iloc[112845,1].split('\n')[i].split(','))
df_1 = pd.DataFrame(df_1_list, columns=df.columns[:-1])

df_2_list = []
for i in range(1, len(df.iloc[142235,1].split('\n'))):
    df_2_list.append(df.iloc[142235,1].split('\n')[i].split(','))
df_2 = pd.DataFrame(df_2_list, columns=df.columns[:-1])

for i in range(2, len(df_1.columns)):
    df_1.iloc[-1,i]=df.iloc[112845,i]
for i in range(2, len(df_2.columns)):
    df_2.iloc[-1,i]=df.iloc[142235,i]

for i in range(len(df.columns)-2, 0, -1):
    df.iloc[142235,i]=df.iloc[142235,1].split('\n')[0].split(',')[i-1]
for i in range(len(df.columns)-2, 0, -1):
    df.iloc[112845,i]=df.iloc[112845,1].split('\n')[0].split(',')[i-1]
```

```python
#Concating the 2 unnested data sets created
df_3=pd.concat([df_1,df_2],ignore_index=True)
#Creating a data set with unnesting the original data set
df_4=pd.concat([df,df_3],ignore_index=True)
```

## Statistical Analysis

```python
#Imporing of statistical testing packages
from scipy.stats import shapiro, levene, f_oneway, kruskal, ttest_ind, pearsonr, spearmanr
```

```python
#Shapiro Test
# H0: Data is Gaussian
# Ha: Data is not Gaussian
test_stat, p_value = shapiro(df_5['price'].sample(1000))
print(test_stat)
print(p_value)
if p_value < 0.05:
    print("Reject H0")
    print("Data is not Gaussian")
else:
    print("Fail to reject H0")
    print("Data is Gaussian")
```

## Price vs AB Test

```python
#Creating the different groups
test_price = df_5.loc[df_5.abtest == 'test', 'price']
control_price = df_5.loc[df_5.abtest == 'control', 'price']
```

```python
#Mean Prices of the groups
df_5.groupby('abtest')['price'].mean()
```

```python
# H0: Control and Test have same average price (high p-value)
# for comapring avg price of two samples (test and control), we use two-sample t test
ttest_ind(test_price, control_price)
```

```python
#Variance Prices of the groups
test_price.var(), control_price.var()
```

```python
# Levene test compares variance
# H0: both have same variance
test_stat, p_value = levene(test_price, control_price)
if p_value < 0.05:
    print("Reject H0")
    print("Variances are different")
else:
    print("Fail to reject H0")
    print("Variances are the same")
```

```python
# Since data from where sample is taken is not Gausian, we need to perform Kruskal-Wallis H-test
# H0: avg price of two groups is same
## Ha: At least one group's avg price is different
f_stat, p_value = kruskal(test_price, control_price)
print(f_stat)
print(p_value)
if p_value<0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

## Price Prediction

The final step would be to create a model to predict the values of price if we were given the features which price is dependent on like vehichle type, gear box, fuel type and notrepaireddamage.

```python
#creating a dataframe with the unique values of the festures
df_6 = df_5.groupby(['vehicletype','gearbox','fueltype','notrepaireddamage'])['name'].count().reset_index().drop(['name'], axis=1)
```

```python
#Using Boot Strap to fetch mean using 95% CI
bootstrapped_means_survey_1_final = []
for i in range(len(df_6)):
    xyz = list(df_6.loc[i,['vehicletype','gearbox','fueltype','notrepaireddamage']])
    survey_1 = df_5.loc[(df_5.vehicletype==xyz[0]) & (df_5.gearbox==xyz[1]) & (df_5.fueltype==xyz[2]) & (df_5.notrepaireddamage==xyz[3]), 'price']
    bootstrapped_means_survey_1 = []
    for reps in range(1000):
        bootstrapped_samples = np.random.choice(survey_1, size=30)
        bootstrapped_mean = np.mean(bootstrapped_samples)
        bootstrapped_means_survey_1.append(bootstrapped_mean)
    df_6.loc[i, ['min_price']] = np.percentile(bootstrapped_means_survey_1, [2.5])
    df_6.loc[i, ['max_price']] = np.percentile(bootstrapped_means_survey_1, [97.5])
    bootstrapped_means_survey_1_final.append(bootstrapped_means_survey_1)
```
